第3題: 請設計一個計算梯形面積的程式，面積計算至小數點後兩位，輸入輸出需與下圖一致。

截圖 2022-05-02 下午2.55.32.png



```python
上底 = int(input('請輸入梯形上底長度:'))
下底 = int(input('請輸入梯形下底長度:'))
高度 = int(input('請輸入梯形的高度:'))
面積 = (上底+下底)*高度/2
print('梯形上底長度為:%d公分，梯形下底長度為:%d公分，梯形高度為:%d，梯形面積為:%0.2f平方公分' %(上底,下底,高度,面積))
```

第4題:請設計一個計算BMI的的程式，輸入輸出結果需如下圖。

截圖 2022-05-02 下午3.16.35.png

```python
height=float(input("請輸入您的身高(cm):請輸入您的身高(cm):"))
weight=float(input("請輸入您的身高(cm):請輸入您的體重(kg):"))
bmi=weight/(height/100)**2
print('身高%d 公分,體重%d公斤,BMI值為%.2f'%(height,weight,bmi))
```

# 第6題: 請設計一個判斷體溫的程式，當體溫高於37.5或小於36，顯示 “體溫異常，請重新量測!”，體溫介於36~37.5間顯示"體溫正常，請進校園!"

截圖 2022-05-02 下午3.04.39.png

截圖 2022-05-02 下午3.08.13.png

```python
體溫 = float(input('請輸入你的體溫'))
if 體溫 >= 36 and 體溫 <= 37.5:
  print("體溫正常，請進校園!")
else:
  print("體溫異常，請重新測量!")
```

```python
體溫 = float(input('請輸入你的體溫'))
if 體溫 >= 36 and 體溫 <= 37.5:
  print("體溫正常，請進校園!")
else:
  print("體溫異常，請重新測量!")
```
