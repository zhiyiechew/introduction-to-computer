### 第五題:
 請用range設計一個程式，使用者只要輸入一個正整數，程式就會列出由 1 加總到該整數，並印出過程
image.png 

```python
sum=0
test=int(input("請輸入一個正整數"))
for m in range(1,test+1):
  sum=sum+m
  print("1到"+str(m)+"的整數和為:",sum)
```
